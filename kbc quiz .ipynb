```python
print("HELLO! WELCOME TO KBC.")
print("PLEASE ENTER YOUR NAME: ")
name=input()
print("THANK YOU FOR PLAYING HERE THIS WONDERFUL GAME \"KBC\"!",name.upper())
print("SO LETS START THE KBC!!!")
qus=["Q1. What is the best mountain in INDIA?\n(A) kashmir\n(B) himachal\n(C) himalaya\n(D) nepal",
     "Q2. How many teams are there in IPL?\n(A) 6\n(B) 7\n(C) 8\n(D) 9",
     "Q3. Who is the prime minister of india?\n(A) Modiji\n(B) Yogiji\n(C) Shushmita\n(D) Niramala",
     "Q4. In which place shree ram born?\n(A) Lanka\n(B) Mathura\n(C) Ayodhya\n(D) Gokul",
     "Q5. Which team wins the IPL trophy?\n(A) Gujarat titens\n(B) CSK\n(C) kolkata\n(D) dilhi Capitals"]
k=0
prize=0
s=["C","B","A","C","B"]
s1=["c","b","a","c","b"]
s3=["himalaya","7","Modiji","Ayodhya","CSK"]
for i in qus:
    print(i)
    responce=input("entre the answer: ")
    if (responce==s[k] or responce==s1[k] or responce==s3[k]):
        print("CONGRETULATION",name,"YOUR ANSWER IS CORRECT")
        prize=(prize+1000)*2
        print("YOUR CURRENT WINING PRIZE IS:",prize,"/-")
        k=k+1
        print("\n")
    else:
        print("SORRY! YOUR ANSWER IS WRONG, GOOD LUCK FOR NEXT TIME",name.upper())
        print("\n")
        break
if(prize==0):
    print("WE ARE SO SORRY TO INFORM YOU THAT YOU WIN NO CASH PRIZE TODAY!")
else:
 print("CONGRATS",name.upper(),"YOU WON A TOTAL AMOUNT OF RS.",prize)

```
